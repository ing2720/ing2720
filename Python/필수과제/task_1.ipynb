- 입력 중 숫자가 아닌 값이 입력되면 ValueError를 처리
- 숫자 외 값이 입력된 경우 어떤 값이 잘못 입력되었는지 사용자에게 알려주는 문구 출력<br>
  (예: f"잘못된 값: {변수}. 숫자를 입력해주세요.")



````python
numbers = []
while True:
    try:
        line = input("숫자를 입력하세요 ('done'이면 종료): ")
        if line == 'done':
            break
        numbers.append(int(line))
    except ValueError:
        print(f"잘못된 값: {line}. 숫자를 입력해주세요.")
````

requests 모듈을 사용하여, 웹 페이지의 HTML을 가져오는 간단한 웹 스크레이퍼를 작성하세요.

'https://kream.co.kr/'에서<br>
데이터를 가져와서 상태 코드와 HTML의 첫 100자를 출력하세요.<br>
(실습 전 requests 모듈을 설치해야 함)

````python
import requests
url = 'https://kream.co.kr/'

try:
    response = requests.get(url)
    response.raise_for_status()
    print(f"상태코드: {response.status_code}")
    html_content = response.text
    print("HTML의 첫 100자:")
    print(html_content[:100])

except requests.exceptions.RequestException as e:
    print(f"오류 발생: {e}")
````

[문제 7]

함수 실행 시간을 측정하는 데코레이터 timer_decorator를 작성하세요.

임의의 함수가 실행되는 시간을 측정하여 "함수 실행 시간: x.xx초" 형태로 출력하세요.

````python
import time

def timer_decorator(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        print(f"함수 실행 시간: {end_time - start_time:.2f}초")
        return result
    return wrapper

@timer_decorator
def example_function():
    sum = 0
    for i in range(1000000):
        sum += i
    return sum

example_function()
````

[문제 8]

여러분이 자주 사용하는 수학적 연산<br>
(예: 삼각형 넓이, 원의 넓이, 직육면체의 넓이)을 수행하는 함수들을 포함하는 mymath.py라는 이름의 모듈을 만드세요.

이 모듈을 사용하여 작성된 함수들을 실제로 임포트하고 사용하는 간단한 프로그램을 작성하세요.

````python
%%writefile mymath.py
import math

def triangle_area(base, height): #삼각형의 넓이 계산
  return 0.5 * base * height

def circle_area(radius): #원의 넓이 계산
  return math.pi * (radius ** 2)

def rectangular_prism_area(length, width, height): #직육면체의 넓이 계산
  return 2 * (length * width + length * height + width * height)

if __name__ == "__main__":
  print("mymath 모듈 테스트:")
  print(f"삼각형 넓이 (밑변=10, 높이=5): {triangle_area(10, 5)}")
  print(f"원의 넓이 (반지름=7): {circle_area(7):.2f}")
  print(f"직육면체 겉넓이 (가로=3, 세로=4, 높이=5): {rectangular_prism_area(3, 4, 5)}")
````

````python
import mymath

삼각형_밑변 = 12
삼각형_높이 = 8
삼각형_넓이 = mymath.triangle_area(삼각형_밑변, 삼각형_높이)
print(f"밑변 {삼각형_밑변}, 높이 {삼각형_높이}인 삼각형의 넓이: {삼각형_넓이}")

원의_반지름 = 10
원의_넓이 = mymath.circle_area(원의_반지름)
print(f"반지름 {원의_반지름}인 원의 넓이: {원의_넓이:.2f}")

직육면체_가로 = 5
직육면체_세로 = 6
직육면체_높이 = 7
직육면체_겉넓이 = mymath.rectangular_prism_area(직육면체_가로, 직육면체_세로, 직육면체_높이)
print(f"가로 {직육면체_가로}, 세로 {직육면체_세로}, 높이 {직육면체_높이}인 직육면체의 겉넓이: {직육면체_겉넓이}")
````

animals라는 패키지를 만들고, 이 패키지 내에 mammals, birds라는 두 개의 서브 모듈을 생성하세요.

- 각 모듈에는 최소한 한 개 이상의 동물에 관한 클래스를 정의하세요<br>
  (예: mammals에는 Dog, birds에는 Eagle)
- 이 패키지와 모듈을 사용하여 동물들의 정보를 출력하는 프로그램을 작성

폴더 및 파일 구조
```
animals/(폴더)
├── __init__.py
├── mammals.py
└── birds.py
```


````python
!mkdir animals
!touch animals/__init__.py
!touch animals/mammals.py
!touch animals/birds.py
````

````python
%%writefile animals/mammals.py
class Dog:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return f"{self.name} says Woof!"

    def info(self):
        return f"나는 포유류인 개입니다. 이름은 {self.name}입니다."
````

````python
%%writefile animals/birds.py
class Eagle:
    def __init__(self, name):
        self.name = name

    def fly(self):
        return f"{self.name} is soaring through the sky!"

    def info(self):
        return f"나는 조류인 독수리입니다. 이름은 {self.name}입니다."
````

````python
from animals import mammals
from animals import birds

my_dog = mammals.Dog("Buddy")
print(my_dog.info())
print(my_dog.speak())

my_eagle = birds.Eagle("Goldy")
print(my_eagle.info())
print(my_eagle.fly())
````
