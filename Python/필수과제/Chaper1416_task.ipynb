__[문제4]__

행렬의 덧셈은 행과 열의 크기가 같은 두 행렬의 같은 행, 같은 열의 값을 서로 더한 결과가 됩니다.

 행렬 arr1과 arr2가 주어졌을 때, 행렬 덧셈의 결과를 출력하는 코드를 작성해주세요(result와 같은 결과가 출력되도록 작성)

arr1 = [[1,2],[2,3]]

arr2 = [[3,4],[5,6]]

result = [[4,6],[7,9]]

```python
arr1 = [[1,2],[2,3]]
arr2 = [[3,4],[5,6]]
result = []
for _ in range(len(arr1)):
    result.append([])
for i in range(len(arr1)):
    for j in range(len(arr1[0])):
        result[i].append(arr1[i][j] + arr2[i][j])
print(result)
```

__[문제5]__

양의 정수 x가 하샤드 수이려면 x의 자릿수의 합으로 x가 나누어져야 합니다.   
예를 들어 18의 자릿수 합은 1+8=9이고, 18은 9로 나누어 떨어지므로 18은 하샤드 수입니다.   
주어진 x를 이용해 x가 하샤드 수인지 아닌지 검사하는 코드를 작성해주세요

x = 17

----

입출력 예시
x	결과값  
10	true  
12	true  
11	false  
13	false  

```python
x = input("하샤드 수를 확인할 숫자를 적어주세요 : ")
x = int(x)
sum_a = 0
for i in str(x):
    sum_a += int(i)
if x % sum_a == 0:
    print("True")
else:
    print("False")
```

String형 배열 seoul의 element중 "Kim"의 인덱스 x를 찾아, "김서방은 x에 있다"는 String을 출력하는 코드를 작성해주세요.   

seoul에 "Kim"은 오직 한 번만 나타나며 잘못된 값이 입력되는 경우는 없습니다.

seoul = ["Jane", "Kim"]

__[문제7]__

0부터 9까지의 숫자 중 일부가 들어있는 정수 배열 numbers가 주어집니다.   
numbers에서 찾을 수 없는 0부터 9까지의 숫자를 모두 찾아 더한 수를 출력해주세요

numbers = [5,8,1,0,6,9]

```python
numbers = [5,8,1,0,6,9]
missing = []
for num in range(10):
    if num in numbers:
        print(f"{num}은 numbers에 있습니다.")
    else:
        print(f"{num}은 numbers에 없습니다.")
        missing.append(num)
print(f"numbers에 없는 숫자들은 {missing} 입니다.")
missing_sum = sum(missing)
print(f"찾아낸 숫자들 더하기: {missing_sum}")
```
