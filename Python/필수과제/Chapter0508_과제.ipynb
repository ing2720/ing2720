__[문제 5: 레벨업]__

모든 캐릭터가 사냥을 성공적으로 마쳤습니다. 이제 경험치를 얻어 레벨업을 해야 합니다.


각 캐릭터의 경험치가 100 이상인지 print()를 이용해 확인해주세요.

<br>

경험치는 아래와 같습니다.

마법사: 120

전사: 150

도적: 110

힐러: 90



```python
aexp = 120
bexp = 150
cexp = 110
dexp = 90

characters = {
    "마법사" : aexp,
    "전사" : bexp,
    "도적" : cexp,
    "힐러" : dexp
}

for name, exp in characters.items():
    if exp >= 100:
        print(f"{name}: 레벨업 성공!")

    else:
        print(f"{name}: 레벨업에 실패하셨습니다. (경험치:{exp})")
```
