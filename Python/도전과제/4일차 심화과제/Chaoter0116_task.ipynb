__[문제 3: 붕어빵 주문 처리]__

손님의 주문을 처리하여 붕어빵 재고를 업데이트하는 기능을 구현하세요.

요구사항:
- 손님의 주문 내용을 기반으로 재고를 업데이트합니다.
- 재고가 부족할 경우, 손님에게 재고 부족을 알리고 재고를 감소시키지 않습니다.
- 재고가 충분할 경우, 주문한 만큼 재고를 감소시키고 판매를 완료합니다.
- 판매가 완료된 경우 판매된 붕어빵 맛과 개수를 출력하세요.

```python
재고 = {
    "팥붕어빵": 10,
    "슈크림붕어빵": 8,
    "초코붕어빵": 5
}
주문내역_str = {
    "팥붕어빵": input("구매하실 팥붕어빵의 갯수를 적어주세요 :"),
    "슈크림붕어빵": input("구매하실 슈크림붕어빵의 갯수를 적어주세요 :"),
    "초코붕어빵": input("구매하실 초코붕어빵의 갯수를 적어주세요:")
}
주문내역 = {}
for 붕어빵_종류, 개수_str in 주문내역_str.items():
    주문내역[붕어빵_종류] = int(개수_str)
재고_충분 = True
판매된_상품 = {}
for 붕어빵_종류, 개수 in 주문내역.items():
    if 붕어빵_종류 in 재고:
        if 재고[붕어빵_종류] >= 개수:
            재고[붕어빵_종류] -= 개수
            판매된_상품[붕어빵_종류] = 개수
        else:
            print(f"{붕어빵_종류}이 부족합니다.")
            재고_충분 = False
if 재고_충분:
    for 종류, 개수 in 판매된_상품.items():
        print(f"{종류} : {개수}")
print("현재 남은 재고",재고)
```

__[문제 4: 여러 손님 주문 처리]__

여러 손님들의 주문을 처리할 수 있도록 시스템을 확장하세요.

요구사항:
- 손님이 "종료"를 입력할 때까지 주문을 계속 받습니다.
- 각 주문마다 재고를 업데이트합니다.
- 현재 재고를 화면에 출력합니다.

```python
재고 = {
    "팥붕어빵": 10,
    "슈크림붕어빵": 8,
    "초코붕어빵": 5
}
주문번호 = 0
print("붕어빵 자동 판매 시스템")
print("주문 입력 시 '종료'를 입력하면 프로그램을 종료합니다.")

while True:
    print("\n현재 재고:", 재고)
    print("\n주문할 붕어빵 종류와 개수를 입력하세요.")
    주문_팥붕어빵_str = input("팥붕어빵 갯수 (종료 입력 시 종료): ")
    if 주문_팥붕어빵_str == "종료":
        break

    주문_슈크림붕어빵_str = input("슈크림붕어빵 갯수 (종료 입력 시 종료): ")
    if 주문_슈크림붕어빵_str == "종료":
        break

    주문_초코붕어빵_str = input("초코붕어빵 갯수 (종료 입력 시 종료): ")
    if 주문_초코붕어빵_str == "종료":
        break
    try:
        주문내역 = {
            "팥붕어빵": int(주문_팥붕어빵_str),
            "슈크림붕어빵": int(주문_슈크림붕어빵_str),
            "초코붕어빵": int(주문_초코붕어빵_str)
        }
        재고_충분 = True
        판매된_상품 = {}
        for 붕어빵_종류, 주문_개수 in 주문내역.items():
            if 붕어빵_종류 in 재고:
                if 재고[붕어빵_종류] < 주문_개수:
                    재고_충분 = False
                    print(f"재고가 부족합니다: {붕어빵_종류} (현재 재고: {재고[붕어빵_종류]}, 주문량: {주문_개수})")
                    break
            else:
                print(f"제공되지 않는 붕어빵 종류입니다: {붕어빵_종류}")
                재고_충분 = False
        if 재고_충분:
            주문번호 += 1
            print(f"{주문번호}번 고객님 주문이 완료되었습니다.")
            for 붕어빵_종류, 주문_개수 in 주문내역.items():
                if 주문_개수 > 0:
                    재고[붕어빵_종류] -= 주문_개수
                    print(f"{주문번호} 고객님 {붕어빵_종류} : {주문_개수}개 나왔습니다")
            print("\n업데이트된 재고:", 재고)
        else:
            print("주문이 취소되었습니다.")
    except ValueError:
        print("올바른 숫자를 입력해주세요.")
print("\n붕어빵 판매 시스템을 종료합니다.")
```

__[문제 5: 관리자 모드]__

관리자 모드를 통해 붕어빵 재고를 추가할 수 있는 기능을 구현하세요.

요구사항:
- 관리자는 특정 붕어빵의 재고를 추가할 수 있습니다.
- "종료"를 입력하면 관리자 모드를 종료합니다.
- 추가된 재고를 업데이트하고 현재 재고를 화면에 출력합니다.

```python
재고 = {
    "팥붕어빵": 10,
    "슈크림붕어빵": 8,
    "초코붕어빵": 5
}

print("\n--- 관리자 모드 ---")
print("재고 관리할 붕어빵 종류와 작업(추가/차감) 및 개수를 입력하세요.")
print("형식: [종류] [추가/차감] [개수]")
print("예시: 팥붕어빵 추가 10")
print("예시: 슈크림붕어빵 차감 3")
print("종료하려면 '종료'를 입력하세요.")

while True:
    print("\n현재 재고:", 재고)
    입력_문자열 = input("입력: ")

    if 입력_문자열.lower() == "종료":
        break

    입력_리스트 = 입력_문자열.split()

    if len(입력_리스트) == 3:
        붕어빵_종류 = 입력_리스트[0]
        작업_종류 = 입력_리스트[1].lower()
        개수_str = 입력_리스트[2]

        try:
            개수 = int(개수_str)

            if 붕어빵_종류 in 재고:
                if 작업_종류 == "추가":
                    if 개수 >= 0:
                        재고[붕어빵_종류] += 개수
                        print(f"{붕어빵_종류} 재고 {개수}개 추가 완료.")
                        print("업데이트된 재고:", 재고)
                    else:
                        print("추가할 개수는 음수일 수 없습니다.")

                elif 작업_종류 == "차감":
                    if 개수 >= 0:
                        if 재고[붕어빵_종류] >= 개수:
                            재고[붕어빵_종류] -= 개수
                            print(f"{붕어빵_종류} 재고 {개수}개 차감 완료.")
                            print("업데이트된 재고:", 재고)
                        else:
                            print(f"{붕어빵_종류} 재고가 부족하여 차감할 수 없습니다. (현재 재고: {재고[붕어빵_종류]}, 차감 요청: {개수})")
                    else:
                         print("차감할 개수는 음수일 수 없습니다.")

                else:
                    print("잘못된 작업 종류입니다. '추가' 또는 '차감'을 입력하세요.")

            else:
                print(f"알 수 없는 붕어빵 종류입니다: {붕어빵_종류}")

        except ValueError:
            print("잘못된 입력 형식입니다. 개수는 숫자로 입력해야 합니다.")

    else:
        print("잘못된 입력 형식입니다. '종류 작업 개수' 또는 '종료' 형식으로 입력하세요.")

print("\n--- 관리자 모드 종료 ---")
```

    - "초코붕어빵": 0

```python
재고 = {
    "팥붕어빵": 10,
    "슈크림붕어빵": 8,
    "초코붕어빵": 5
}

가격 = {
    "팥붕어빵": 1000,
    "슈크림붕어빵": 1200,
    "초코붕어빵": 1500
}

판매_개수 = {
    "팥붕어빵": 0,
    "슈크림붕어빵": 0,
    "초코붕어빵": 0
}

주문번호 = 0
관리자_비밀번호 = "2939"
시스템_종료_요청 = False

print("붕어빵 자동 판매 시스템")

while not 시스템_종료_요청:
    print("\n--- 모드 선택 ---")
    print("1. 손님 모드")
    print("2. 관리자 모드")
    print("3. 시스템 종료 (매출 확인)")

    모드_선택 = input("원하는 모드를 선택하세요 (1, 2, 3): ")

    if 모드_선택 == "1":
        print("\n--- 손님 모드 ---")
        while True:
            print("\n현재 재고:", 재고)
            print("\n주문할 붕어빵 종류와 개수를 입력하세요.")
            주문_팥붕어빵_str = input("팥붕어빵 갯수 (종료 입력 시 모드 종료): ")
            if 주문_팥붕어빵_str.lower() == "종료":
                break

            주문_슈크림붕어빵_str = input("슈크림붕어빵 갯수 (종료 입력 시 모드 종료): ")
            if 주문_슈크림붕어빵_str.lower() == "종료":
                break

            주문_초코붕어빵_str = input("초코붕어빵 갯수 (종료 입력 시 모드 종료): ")
            if 주문_초코붕어빵_str.lower() == "종료":
                break

            try:
                주문내역 = {
                    "팥붕어빵": int(주문_팥붕어빵_str),
                    "슈크림붕어빵": int(주문_슈크림붕어빵_str),
                    "초코붕어빵": int(주문_초코붕어빵_str)
                }
                재고_충분 = True
                for 붕어빵_종류, 주문_개수 in 주문내역.items():
                    if 붕어빵_종류 in 재고:
                        if 재고[붕어빵_종류] < 주문_개수:
                            재고_충분 = False
                            print(f"재고가 부족합니다: {붕어빵_종류} (현재 재고: {재고[붕어빵_종류]}, 주문량: {주문_개수})")
                            break
                    else:
                        재고_충분 = False
                        print(f"제공되지 않는 붕어빵 종류입니다: {붕어빵_종류}")
                        break

                if 재고_충분:
                    주문번호 += 1
                    print(f"{주문번호}번 고객님 주문이 완료되었습니다.")
                    for 붕어빵_종류, 주문_개수 in 주문내역.items():
                        if 주문_개수 > 0:
                            재고[붕어빵_종류] -= 주문_개수
                            판매_개수[붕어빵_종류] += 주문_개수
                            print(f"{주문번호} 고객님 {붕어빵_종류} : {주문_개수}개 나왔습니다")
                    print("\n업데이트된 재고:", 재고)
                else:
                    print("주문이 취소되었습니다.")
            except ValueError:
                print("올바른 숫자를 입력해주세요.")
        print("\n--- 손님 모드 종료 ---")

    elif 모드_선택 == "2":
        print("\n--- 관리자 모드 진입 ---")
        입력_비밀번호 = input("관리자 비밀번호를 입력하세요: ")

        if 입력_비밀번호 == 관리자_비밀번호:
            print("비밀번호 확인 완료.")
            print("\n--- 관리자 모드 ---")
            print("재고 관리할 붕어빵 종류와 작업(추가/차감) 및 개수를 입력하세요.")
            print("형식: [종류] [추가/차감] [개수]")
            print("예시: 팥붕어빵 추가 10")
            print("예시: 슈크림붕어빵 차감 3")
            print("손님모드나 매출표로 전환하려면 '전환'을 입력하세요.")

            while True:
                print("\n현재 재고:", 재고)
                입력_문자열 = input("입력: ")

                if 입력_문자열.lower() == "전환":
                    break

                입력_리스트 = 입력_문자열.split()

                if len(입력_리스트) == 3:
                    붕어빵_종류 = 입력_리스트[0]
                    작업_종류 = 입력_리스트[1].lower()
                    개수_str = 입력_리스트[2]

                    try:
                        개수 = int(개수_str)

                        if 붕어빵_종류 in 재고:
                            if 작업_종류 == "추가":
                                if 개수 >= 0:
                                    재고[붕어빵_종류] += 개수
                                    print(f"{붕어빵_종류} 재고 {개수}개 추가 완료.")
                                    print("업데이트된 재고:", 재고)
                                else:
                                    print("추가할 개수는 음수일 수 없습니다.")

                            elif 작업_종류 == "차감":
                                if 개수 >= 0:
                                    if 재고[붕어빵_종류] >= 개수:
                                        재고[붕어빵_종류] -= 개수
                                        print(f"{붕어빵_종류} 재고 {개수}개 차감 완료.")
                                        print("업데이트된 재고:", 재고)
                                    else:
                                        print(f"{붕어빵_종류} 재고가 부족하여 차감할 수 없습니다. (현재 재고: {재고[붕어빵_종류]}, 차감 요청: {개수})")
                                else:
                                     print("차감할 개수는 음수일 수 없습니다.")

                            else:
                                print("잘못된 작업 종류입니다. '추가' 또는 '차감'을 입력하세요.")

                        else:
                            print(f"알 수 없는 붕어빵 종류입니다: {붕어빵_종류}")

                    except ValueError:
                        print("잘못된 입력 형식입니다. 개수는 숫자로 입력해야 합니다.")

                else:
                    print("잘못된 입력 형식입니다. '종류 작업 개수' 또는 '종료' 형식으로 입력하세요.")
            print("\n--- 관리자 모드 종료 ---")

            while True:
                print("\n관리자 모드가 종료되었습니다.")
                print("1. 손님 모드로 전환")
                print("2. 시스템 종료")
                종료_선택 = input("선택: ")

                if 종료_선택 == "1":
                    break
                elif 종료_선택 == "2":
                    시스템_종료_요청 = True
                    break
                else:
                    print("잘못된 선택입니다. 1 또는 2를 입력하세요.")
        else:
            print("비밀번호가 틀렸습니다. 관리자 모드 진입에 실패했습니다.")

    elif 모드_선택 == "3":
        시스템_종료_요청 = True

    else:
        print("잘못된 모드 선택입니다. 1, 2 또는 3을 입력하세요.")

if 시스템_종료_요청:
    print("\n붕어빵 판매 시스템을 종료합니다.")

    총_매출 = 0
    print("\n--- 매출 확인 ---")
    for 붕어빵_종류, 판매된_개수 in 판매_개수.items():
        if 붕어빵_종류 in 가격:
            품목_매출 = 판매된_개수 * 가격[붕어빵_종류]
            print(f"{붕어빵_종류}: {판매된_개수}개 판매, 매출: {품목_매출}원")
            총_매출 += 품목_매출

    print(f"\n총 매출: {총_매출}원")
```
