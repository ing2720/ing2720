__[문제 2: 입금]__

가지고 있는 돈을 입금합니다.

입금 금액을 입력받고, 잔액을 갱신한 후 새로운 잔액을 출력하세요. 입금 거래를 영수증 리스트에 기록하세요.

- 입금 금액은 input()을 이용해 받습니다.
- 영수증은 list()로 선언합니다.
- 영수증에 들어가는 정보는 튜플 형태입니다.
- 튜플에 담기는 정보는 (입금 or 출금, 입금 or 출금 금액, 현재 잔액)순 입니다.

(변수명 예시: 영수증 리스트 - receipts, 입금 금액 - deposit_amount)


```python
# input()을 사용해서 입금 금액을 사용자로부터 입력받아 보세요.
# 입력받은 값은 숫자(int)로 바꿔야 합니다.

# 잔액(balance)에 입금 금액을 더해서 잔액을 갱신해 보세요.

# receipts라는 리스트를 생성하고, 거래 정보를 튜플 형태로 추가해 보세요.

# 튜플에는 (입금/출금, 금액, 현재 잔액)을 저장하세요.

# 마지막으로 잔액을 출력해 보세요.
잔액 = 1000
bank = input("입금하실건지 출금하실건지 말해주세요")
if bank == "입금":
    print("입금을 진행합니다")
    입금금액 = int(input("입금하실 금액을 입력해 주십시오"))
    입금후금액 = 잔액 + 입금금액
    입금내역 = (bank, 입금금액, 입금후금액)
    입금일까출금일까 = 입금내역[0]
    들어온금액 = 입금내역[1]
    잔액 = 입금내역[2]
    print(f"{입금일까출금일까} : {들어온금액}원, 잔액 :{잔액}원")
elif bank == "출금":
    print("출금을 진행합니다")
    출금금액 = int(input("출금하실 금액을 입력해 주십시오"))
    출금후금액 = 잔액 - 출금금액
    if 잔액 >= 출금금액:
            출금내역 = (bank, 출금금액, 출금후금액)
            출금일까입금일까 = 출금내역[0]
            나간금액 = 출금내역[1]
            잔액 = 출금내역[2]
            print(f"{출금일까입금일까} : {나간금액}원, 잔액 : {잔액}원")

    else:
        print(f"잔액이 부족합니다. 현재잔액은 {잔액}원 입니다.")
else:
    print("다시 입력해주세요")
```

__[문제 3: 출금]__

사용자가 돈을 출금합니다. 출금 금액을 입력받고, 잔액을 갱신한 후 새로운 잔액을 출력하세요. 출금 거래를 영수증 리스트에 기록하세요.

- 출금 금액은 input()을 이용해 받습니다.
- 현재 잔액보다 출금 요청 금액이 많을 경우 현재 잔액 만큼만 출금되도록 합니다.
- 영수증에 들어가는 정보는 튜플 형태입니다.
- 튜플에 담기는 정보는 (입금 or 출금, 입금 or 출금 금액, 현재 잔액)순 입니다.

(변수명 예시: 출금금액 - withdraw_amount)

```python
# input()을 사용해서 출금 금액을 입력받아 보세요.

# 입력받은 금액이 잔액보다 크다면, 출금 금액을 잔액으로 바꾸세요.

# 잔액(balance)에서 출금 금액을 빼고 갱신해 보세요.

# receipts 리스트에 출금 거래 정보를 튜플로 추가해 보세요.

# 마지막으로 현재 잔액을 출력해 보세요.
잔액 = 1000
bank = input("입금하실건지 출금하실건지 말해주세요")
if bank == "입금":
    print("입금을 진행합니다")
    입금금액 = int(input("입금하실 금액을 입력해 주십시오"))
    입금후금액 = 잔액 + 입금금액
    입금내역 = (bank, 입금금액, 입금후금액)
    입금일까출금일까 = 입금내역[0]
    들어온금액 = 입금내역[1]
    잔액 = 입금내역[2]
    print(f"{입금일까출금일까} : {들어온금액}원, 잔액 :{잔액}원")
elif bank == "출금":
    print("출금을 진행합니다")
    출금금액 = int(input("출금하실 금액을 입력해 주십시오"))
    출금후금액 = 잔액 - 출금금액
    if 잔액 >= 출금금액:
            출금내역 = (bank, 출금금액, 출금후금액)
            출금일까입금일까 = 출금내역[0]
            나간금액 = 출금내역[1]
            잔액 = 출금내역[2]
            print(f"{출금일까입금일까} : {나간금액}원, 잔액 : {잔액}원")

    else:
        print(f"잔액이 부족합니다. 현재잔액은 {잔액}원 입니다.")
else:
    print("다시 입력해주세요")
```

__[문제 4: 모든 거래 내역 확인]__

모든 거래 내역(영수증)을 요청받았을 때 출력하세요.
- 지금까지 영수증에 추가된 정보를 출력합니다.
- 출력 형태는 자유입니다.

```python
# receipts 리스트에 저장된 모든 거래를 하나씩 출력해 보세요.
# for문을 이용해서 각 거래 정보를 반복하며 출력하면 됩니다.
# 출력 형식은 자유롭게 구성해도 좋아요.

잔액 = 1000
receipts = []
```

```python
bank = input("입금하실건지 출금하실건지 말해주세요")
if bank == "입금":
    print("입금을 진행합니다")
    try:
        입금금액 = int(input("입금하실 금액을 입력해 주십시오"))
        입금후금액 = 잔액 + 입금금액
        입금내역 = (bank, 입금금액, 입금후금액)
        receipts.append(입금내역)
        거래종류 = 입금내역[0]
        들어온금액 = 입금내역[1]
        잔액 = 입금내역[2]
        print(f"{거래종류} : {들어온금액}원, 잔액 :{잔액}원")
        잔액 = 입금후금액
    except ValueError:
        print("숫자를 입력해주세요")
elif bank == "출금":
    print("출금을 진행합니다")
    try:
        출금금액 = int(input("출금하실 금액을 입력해 주십시오"))
        출금후금액 = 잔액 - 출금금액
        if 잔액 >= 출금금액:
            출금내역 = (bank, 출금금액, 출금후금액)
            receipts.append(출금내역)
            거래종류 = 출금내역[0]
            나간금액 = 출금내역[1]
            잔액 = 출금내역[2]
            print(f"{거래종류} : {나간금액}원, 잔액 : {잔액}원")
            잔액 = 출금후금액
    except ValueError:
        print("숫자를 입력해주세요")

    else:
        print(f"잔액이 부족합니다. 현재잔액은 {잔액}원 입니다.")
else:
    print("다시 입력해주세요")
print("\n--- 모든 거래 내역 ---")
if not receipts: # receipts 리스트가 비어있다면 (거래가 없다면)
    print("아직 거래 내역이 없습니다.")
else:
    for receipts_tuple in receipts:
        거래_종류 = receipts_tuple[0]
        금액 = receipts_tuple[1]
        잔액 = receipts_tuple[2]
        print(f"{거래_종류} : {금액}원, 잔액 : {잔액}원")
print("---------------------")
```
